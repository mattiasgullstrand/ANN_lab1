```python
from RBF_Net import RBF_Net
import math
import numpy as np
import matplotlib.pyplot as plt
from single_layer_delta_rule_v2 import Single_Layer_Delta_Rule
from two_layer_perceptron_v2 import Two_Layer_Delta_Rule
```

```python
X_train = np.arange(0, 2*math.pi, 0.1)
X_train_clean = np.copy(X_train)
X_train += np.random.normal(0, 0.1, X_train.shape[0])
X_train = X_train.reshape(1, X_train.shape[0])
X_train_clean = X_train_clean.reshape(1, X_train.shape[0])

X_test = np.arange(0.05, 2*math.pi, 0.1)
X_test_real = np.copy(X_test)
X_test += np.random.normal(0, 0.1, X_test.shape[0])
X_test = X_test.reshape(1, X_test.shape[0])
```

```python
X_train.shape
```

```python
def sin_2(x):
    return np.sin(2*x)
```

```python
def square(x):
    arr = np.zeros(x.shape)
    for i in range(arr.shape[0]):
        if np.sin(x[i]) >= 0:
            arr[i] = 1
        else:
            arr[i] = -1
    return arr
```

```python
f_train = sin_2(X_train) + np.random.normal(0, 0.1, X_train.shape[1])
f_test_real = sin_2(X_test_real)
f_test = f_test_real + np.random.normal(0, 0.1, X_test.shape[1])
```

```python
dim_list = list(range(1, 20))
sigma_list = [0.3, 0.5, 1, 2, 5]
errors = []
for H_dim in dim_list:
    rbf_net = RBF_Net(X_train, f_train.T, H_dim, weight_learn = 'delta', CL = False)
    rbf_net.sigmas *= sigma_list[1]
    rbf_net.eta = 0.5
    rbf_net.epochs = 100
    rbf_net.train()
    f_approx = rbf_net.predict(X_test).flatten()
    test_error = np.mean(np.abs(f_approx - f_test_real))
    errors.append(test_error)
    print("Number of nodes: {}, Error: {}".format(H_dim, test_error))
plt.plot(dim_list, errors)
plt.xlabel("Number of units")
plt.ylabel("Absolute residual error")
plt.title("Sine function")
```

```python
dim_list = list(range(1, 20))
sigma_list = [0.3, 0.5, 1, 2, 5]
errors = []
epoch_list = [50, 100, 150, 200, 250, 300]
for epochs in epoch_list:
    rbf_net = RBF_Net(X_train, f_train.T, 9, weight_learn = 'delta', CL = False)
    rbf_net.sigmas *= sigma_list[1]
    rbf_net.eta = 0.003
    rbf_net.epochs = epochs
    rbf_net.train()
    f_approx = rbf_net.predict(X_test).flatten()
    test_error = np.mean(np.abs(f_approx - f_test_real))
    errors.append(test_error)
    print("Number of epochs: {}, Error: {}".format(epochs, test_error))
plt.plot(epoch_list, errors)
plt.xlabel("Number of units")
plt.ylabel("Absolute residual error")
plt.title("Sine function")
```

```python
rbf_net = RBF_Net(X_train_clean, f_train.T, 12, weight_learn = 'delta', CL = True, init_mode = 'equal')
rbf_net.epochs = 300
rbf_net.sigmas *= 1
rbf_net.eta = 0.01
rbf_net.train()
f_approx = rbf_net.predict(X_test).flatten()
plt.figure(figsize=(9,8))
plt.plot(X_test[0], f_test, 'bo', markersize = 3, label = "Test data")
plt.plot(X_test_real, f_test_real, label = "Real function")
plt.plot(X_test[0], f_approx, 'yo', label = "Approximated function values")
plt.plot(rbf_net.mus, sin_2(rbf_net.mus), 'ro', label = "Unit locations")
plt.legend(loc='upper right')
plt.xlabel("Number of units")
plt.ylabel("Absolute residual error")
plt.title("Sine function")
```

```python
f_sq_train = square(X_train[0]).reshape(f_train.shape)
f_sq_test_real = square(X_test_real)
f_sq_test = square(X_test[0]).reshape(f_train.shape)
```

```python
dim_list = list(range(1, 20))
errors = []
for H_dim in dim_list:
    rbf_net = RBF_Net(X_train, f_sq_train.T, H_dim, weight_learn = 'delta', CL = False)
    rbf_net.epochs = 100
    rbf_net.sigmas *= 1
    rbf_net.eta = 0.001
    rbf_net.train()
    f_sq_approx = rbf_net.predict(X_test).flatten()
    test_error = np.mean(np.abs(f_sq_approx - f_sq_test))
    errors.append(test_error)
    print("Number of nodes: {}, Error: {}".format(H_dim, test_error))
plt.plot(dim_list, errors)
plt.xlabel("Number of units")
plt.ylabel("Absolute residual error")
plt.title("Square function")
```

```python
rbf_net = RBF_Net(X_train, f_sq_train.T, 9, weight_learn = 'delta', CL = True)
rbf_net.epochs = 500
rbf_net.sigmas *= 0.6
rbf_net.eta = 0.01
rbf_net.train()
f_sq_approx = rbf_net.predict(X_test).flatten()
plt.plot(X_test[0], f_sq_test[0])
plt.plot(X_test[0], f_sq_approx)
plt.plot(rbf_net.mus, square(rbf_net.mus), 'bo', color = 'r')
```
